<h1 style="color: #8000ff; font-size: 3rem; text-align: center; font-family: 'Segoe UI', sans-serif; margin-bottom: 20px;">
   Option1 : fill missing data with pandas
</h1>


```python
# fill the make column
cars_missing_data["Make"].fillna("missing" , inplace = True)
# fill the colour column
cars_missing_data["Colour"].fillna("missing" , inplace = True)
#fill the odometor column
cars_missing_data["Odometer (KM)"].fillna(cars_missing_data["Odometer (KM)"].mean() , inplace = True)
#fill door column
cars_missing_data["Doors"].fillna(4 , inplace = True)
```

```python
cars_missing_data.isna().sum()
```

```python
# lets remove rows which price is na
cars_missing_data.dropna(inplace=True)
```

```python
cars_missing_data.isna().sum()
```

```python
X = cars_missing_data.drop("Price" , axis=1)
Y = cars_missing_data["Price"]
```

```python
# turn the categories in number
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_featues = ["Make" , "Colour" , "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot" , one_hot , categorical_featues )] , remainder="passthrough")

transformed_X = transformer.fit_transform(cars_missing_data)
transformed_X
```

<h1 style="color: #8000ff; font-size: 3rem; text-align: center; font-family: 'Segoe UI', sans-serif; margin-bottom: 20px;">
   Option2 : fill missing data with Scikit-Learn
</h1>

```python
car_missing_sk = pd.read_csv("scikit-learn-data/car-sales-extended-missing-data.csv")
car_missing_sk.head()
car_missing_sk.isna().sum()
```

```python
#drop the rows with no labels
car_missing_sk.dropna(subset=["Price"] , inplace= True)
car_missing_sk.isna().sum()
```

```python
x = car_missing_sk.drop("Price",axis = 1)
y = car_missing_sk["Price"]
```

```python
# fill missing values using scikit learn 
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# fill categorical values with missing and numerical values with mean
category_imputer = SimpleImputer(strategy="constant" , fill_value="missing")
door_imputer = SimpleImputer(strategy="constant" , fill_value = 4)
num_imputer = SimpleImputer(strategy="mean")

#define columns
category_features = ["Make","Colour"]
door_features = ["Doors"]
num_features = ["Odometer (KM)"]

#create an imputer (something that fills value)
imputer = ColumnTransformer([("category_imputer" , category_imputer  , category_features) , ("door_imputer" , door_imputer , door_features),
                            ("num_imputer" , num_imputer , num_features)])
#transform the data
filled_x = imputer.fit_transform(x)
filled_x
```

```python
cars_updated_data = pd.DataFrame(filled_x , columns=["Make" , "colour", "Doors", "Odometer (km) "])
```

```python
cars_updated_data.isna().sum()
```

```python
# lets convert the data to numerical data
# turn the categories in number
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ["Make" , "colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")


transformed_x = transformer.fit_transform(cars_updated_data)
transformed_x
```

```python
# we have got our data as num 
#now lets fix the model
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(transformed_x , y , test_size=0.2)
model = RandomForestRegressor()
model.fit(x_train , y_train)
```

```python
model.score(x_test , y_test)
```
