```python
car_sales = pd.read_csv("scikit-learn-data/car-sales-extended.csv")
car_sales.head() 
```

```python
car_sales.dtypes
```

```python
car_sales.head()
```

```python
X = car_sales.drop("Price" , axis=1)
Y = car_sales["Price"]
```

```python
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X,Y)
model.score(X_test , Y_test)
```

```python
# turn the categories in number
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_featues = ["Make" , "Colour" , "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot" , one_hot , categorical_featues )] , remainder="passthrough")

transformed_X = transformer.fit_transform(X)
```

```python
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
# now that we have converted our data in numeric lets refit pur model
np.random.seed(42)
X_train , X_test , Y_train , Y_test = train_test_split(transformed_X , Y , test_size=0.2)
model.fit(X_train , Y_train)
```

```python
model.score(X_test , Y_test)
```
