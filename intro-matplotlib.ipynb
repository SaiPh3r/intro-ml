<h1 style="text-align: center; color:orange">Matplotlib Example Workflow</h1>

```python
# 0)  Import matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

#1) Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

#2) Setup Plot
fig , ax = plt.subplots(figsize=(7,5))

#3) Plot Data
ax.plot(x,y)

#4) Customize Plot
ax.set(title = "Simple-Plot",
        xlabel = "x-axis",
        ylabel = "y-axis")
#5) Save and Show figure
fig.savefig("numpy-images/sample-plot.png")

plt.show()
```

```python
nut_prices = {"almond":10 , "cashew":8, "peanut":12}
nut_prices
```

```python
fig , ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values())
ax.set(title="prices of diff nuts?" , ylabel="$ price")
plt.show()
```

```python
fig , ax = plt.subplots()
ax.barh(nut_prices.keys() , nut_prices.values())
plt.show()
```

```python
# lets see histogram now
fig , ax = plt.subplots()
s = np.random.randn(1000)
ax.hist(s)
plt.show()
```

```python
# lets see some subplots (option 1)
#we will use this option 
fig , ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows=2 , ncols=2 , figsize = (14,8))
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10) , np.random.random(10))
ax3.hist(s)
ax4.bar(nut_prices.keys(), nut_prices.values())
plt.show()
```

```python
#subplots option 2
# will use option 1 
fig , ax = plt.subplots(nrows=2 , ncols=2 , figsize = (14,8))
ax[0,0].plot(x , x/2)
ax[0,1].scatter(np.random.random(10) , np.random.random(10))
ax[1,0].hist(s)
ax[1,1].bar(nut_prices.keys(), nut_prices.values())
plt.show()
```

<h1 style="text-align: center; color:orange">Plotting using OO method</h1>

```python
heart_data = pd.read_csv("heart-disease (1).csv")
heart_data.head(10)
```

```python
over_50 = heart_data[heart_data["age"]>50]
```

```python
over_50.head(10)
```

```python
fig , ax  = plt.subplots(figsize = (10,10))
scatter = ax.scatter(over_50["age"], over_50["chol"], c = over_50["target"] )
ax.set(title = "hear-disease" , xlabel="age" , ylabel="chol")
ax.legend(*scatter.legend_elements() , title = "target")
plt.show()
```

```python
over_50.head()
```

```python
fig , (ax0 , ax1) = plt.subplots(nrows = 2 , ncols=1 , figsize = (10,10) , sharex=True)
scatter = ax0.scatter(over_50["age"] , over_50["chol"] , c=over_50["target"] )
ax0.set(title = "hear-disease" ,  ylabel="chol")
ax0.legend(*scatter.legend_elements() , title = "target")

scatter  = ax1.scatter(over_50["age"] , over_50["thalach"] , c = over_50["target"])
ax1.set(title="age and thalach" , xlabel="age", ylabel="thalach")
ax1.legend(*scatter.legend_elements() , title = "target")
fig.suptitle("heart-disease-analysis")
plt.show()





```
